```python
import pandas as pd
```

```python
df=pd.read_csv("data_clean.csv")
df.head()
```

```python
cols_to_drop = [
    'Unemployment rate',
    'Inflation rate',
    'GDP'
]

df.drop(columns=cols_to_drop, inplace=True)
```

```python
cols_to_drop = [

    'Course',
    'Daytime/evening attendance'
]

df.drop(columns=cols_to_drop, inplace=True)
```

```python
for col in df.columns:
    if df[col].nunique() == 1:
        print("Dropping:", col)
        df.drop(columns=[col], inplace=True)
```

```python
cols_to_drop = [
   'Nationality'
]

df.drop(columns=cols_to_drop, inplace=True)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
